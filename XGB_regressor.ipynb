```python
from sklearn.model_selection import train_test_split
import pandas as pd
import pickle
import numpy as np

import os
os.system('pip install katonic-1.6.1-py3-none-any.whl[ml]')
os.system('pip install plotly')

from katonic.ml.client import set_exp
from katonic.ml.regression import Regressor
import plotly
```

```python
data = pd.read_csv('Feature_engineering.csv')
data
```

```python
# Splitting the Dataset.

X=data.drop('price',axis='columns')
y=data.price
```

#### Split the dataset into training and validation sets

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
exp_name = "house_price_prediction_katonic_sdk_pipeline"
```

```python
artifacts = {
    "data_path": "/home/katonic/house_price_pipelines/Datasets/Bengaluru_House_Data.csv",
}
```

```python
regressor = Regressor(
    X_train,
    X_test,
    y_train,
    y_test,
    experiment_name=exp_name,
    source_name="House_Price_pred_mlflow_sdk_test.ipynb", # /home/katonic/model-cluster-backup/
    features=X_train.columns.to_list(),
    artifacts=artifacts,
)
```

```python
parames = {
   'max_depth':[6,7,10,50,100],
    'eta':.3,
    'subsample': 1,
    'colsample_bytree': 1,

}

xgb_model = regressor.XGBRegressor(is_tune = True, params = parames)
```

```python
pickle_out = open("xgb_regressor.pickle","wb")
pickle.dump(xgb_model, pickle_out)
```
