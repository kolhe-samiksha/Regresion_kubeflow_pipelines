```python
with open('pickle_files/regressor_obj.pickle', 'rb') as f:
    regressor_obj = pickle.load(f)
```

```python
logged_reg_models = regressor_obj.search_runs("32")
logged_reg_models
```

```python
logged_reg_models = regressor_obj.search_runs("32")
rename_logged_models = logged_reg_models.rename(columns = {'metrics.MAE':'MAE','metrics.MSE':'MSE','metrics.R2':'R2','metrics.RMSE':'RMSE','metrics.RMSLE':'RMSLE','tags.experiment_name':'Experiment_name','tags.mlflow.parentRunId':'Parent_run_id'})
best_model_sort_r2 = rename_logged_models.sort_values(['R2'],ascending=False)
best_model_sort_r2 = best_model_sort_r2[:11]
best_model_sort_rmse = best_model_sort_r2.sort_values(['RMSE'],ascending=True).reset_index(drop=True)
best_model = best_model_sort_rmse['run_name'][0]
best_run_id = best_model_sort_rmse['Parent_run_id'][0]
best_model_sort_rmse
```

```python
logged_model_url = best_model_sort_rmse['artifact_uri'][0]
logged_model = f'{logged_model_url}/{best_model}'
logged_model
```

```python
from katonic.ml.client import load_model

model = load_model(logged_model)
```

```python
data = pd.read_csv('Feature_engineering.csv')
X=data.drop('price',axis='columns')
y=data.price

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
preds = model.predict(X_test)
```

```python
y_test_df = pd.DataFrame(y_test).reset_index(drop=True)
y_test_df['price_pred'] = preds
```

```python
y_test_df.to_csv('/predictions.csv')
```
